# Statistics

## t-test

```python
import scipy.stats as stats
```

```python
pvalList = []
for x in range(genes.shape[0]):
    st, pval = stats.ttest_ind(genes.iloc[x, 1:11] , genes.iloc[x, 12:21])
    pvalList.append(pval)
    if(pval < 0.05):
        print(x, genes.iloc[x,0], pval)
```
